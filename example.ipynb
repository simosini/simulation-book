# creating a pseudorandom numbers generator

```python
s = 3
m = 2**31 -1
a = 7**5
```

```python
def generator(seed, module, multiplier, quantity):
    nums = [seed]
    nxt = (a*seed) % m 
    for i in range (quantity):
        nums.append(nxt)
        nxt = (a*nxt) % m
    return nums
```

```python
gen = generator(s, m, a, 10)
gen
```

Let's play with graphs now.
We import some data and we just display them correctly

```python
%matplotlib notebook
import matplotlib.pyplot as plt

x = [1910, 1920, 1930]
y1 = [8,4,14]
y2 = [11,3,15]


plt.plot(x,y1, color='green', label='green')
plt.plot(x,y2, color='red', label='red')
plt.legend(loc='upper center')
plt.axis([1900,1930,0,16])
plt.ylabel('values')
plt.xlabel('year')
plt.show()
```

```python
%matplotlib notebook
t = np.arange(0.,5.,.2)
plt.plot(t,t**2,'r--', label='squared')
plt.plot(t, t , 'bs', label='linear')
plt.plot(t, t**3, 'g^', label='cube')
plt.legend(loc='upper left')
plt.text(2,80, 'different forms', color='red')
plt.grid(True)
plt.xlabel('sample values')
plt.ylabel('function')
plt.title('Figure example')
plt.show()
```

```python
%matplotlib notebook
y = np.random.normal(loc=0.5, scale=0.4,size=1000)
y = y[(y > 0) & (y < 1)] #filtering values keeping only those in the interval (0,1) 
y.sort()
x = np.arange(len(y))
fig, axs = plt.subplots(2,2, sharex=True)

#linear scale
ax = axs[0,0]
ax.plot(x,y)
ax.set_yscale('linear')
ax.set_title('linear')
ax.grid(True)

#log scale
ax = axs[0,1]
ax.plot(x,y)
ax.set_yscale('log')
ax.set_title('log')
ax.grid(True)

plt.show()


```
